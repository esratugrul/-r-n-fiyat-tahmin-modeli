```python
import numpy as np
import pandas as pd
import xgboost as xgb
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import model_selection
```

```python
# Fiyat tahmin etmek istediğimiz için y değişkeni "ürün fiyatı"
```

```python
X = df4.drop(["ürün fiyatı"], axis=1)
Y = df4["ürün fiyatı"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.25, random_state = 42)
```

```python
# GridSearch
```

```python
params = { "colsample_bytree":[0.6,0.8,1.0],
          "learning_rate":[0.01,0.05,0.1],
          "max_depth":[3,5,7],
          "n_estimators":[100,300,500]}
```

```python
xgb = XGBRegressor()
```

```python
grid = GridSearchCV(xgb, params, cv = 10, n_jobs = -1, verbose=2)
```

```python
grid.fit(X_train, Y_train)
```

```python
grid.best_params_
```

```python
xgb1 = XGBRegressor(colsample_bytree = 0.8, learning_rate = 0.05, max_depth = 3, n_estimators = 1000)
```

```python
model_xgb = xgb1.fit(X_train, Y_train)
```

```python
model_xgb.predict(X_test)[0:10]
```
